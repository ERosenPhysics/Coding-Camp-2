```
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
from google.colab import files
data_to_load = files.upload()
```

```
import io
df = pd.read_csv(io.BytesIO(data_to_load['Elevator Ride #1.csv']))
```

```
df.head(25)
```

```
x = df['time']
y = df['aT (m/s^2)']

# plt.plot() makes a line graph
fig = plt.figure(figsize=(15, 10))
plt.plot(x, y)
plt.axhline(0, color='gray')   # try changing the color. It only knows some color names, though.
plt.title('Total Acceleration of my phone in the elevator')
plt.xlabel('Time (s)')
plt.ylabel('m/s/s');
```

```
x2 = df['time']
y2 = df['az (m/s^2)']

# plt.plot() makes a line graph
fig = plt.figure(figsize=(15, 10))
plt.plot(x2, y2)
plt.axhline(0, color='gray')   # try changing the color. It only knows some color names, though.
plt.title('Acceleration of my phone in y in the elevator')
plt.xlabel('Time (s)')
plt.ylabel('m/s/s');
```
