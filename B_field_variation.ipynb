```python
data2=pd.read_excel('https://docs.google.com/spreadsheets/d/1kGmd3B8kBNWlUN9QPZknboCAfEdODyxpt1rpTRX8Rdc/export')
data2.head(5)
```

```python
# caluclate a Holmes t' to compare the two samples
```

```python
abs_mean1 = data1['Absolute field (µT)'].mean()
abs_mean2 = data2['Unnamed: 5'].mean()
print(abs_mean1)
print(abs_mean2)
```

```python
stdev_data1=np.std(data1['Absolute field (µT)'])
stdev_data2=np.std(data2['Unnamed: 5'])
print(stdev_data1)
print(stdev_data2)
```

```python
ldata1=len(data1['Absolute field (µT)'])
ldata2=len(data2['Unnamed: 5'])
print(ldata1)
print(ldata2)
```

```python
sunc1=(stdev_data1)/((ldata1)**1/2)
sunc2=(stdev_data2)/((ldata2)**1/2)
print(sunc1)
print(sunc2)
```

```python
combineduncertainty=(((sunc1)**2) + ((sunc2)**2))**1/2
print(combineduncertainty)
```

```python
holmes=(abs_mean1-abs_mean2)/combineduncertainty
print(holmes)
```
